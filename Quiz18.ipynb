Firstly, I imported pandas and used read_csv to read the Airbnb in NYC data.

```python
import pandas as pa
df = pa.read_csv("https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv")
df.head()
```

Here I am using get_dummies to turn column of neighbourhood_group into separate columns of 0s and 1s. 

```python
pa.get_dummies(df.neighbourhood_group).head()
```

Now, I am using concat to retuen it into the original dataset.

```python
df1 = pa.concat([df,pa.get_dummies(df.neighbourhood_group)],axis = 1)

df1 = df1.drop('neighbourhood_group', axis = 1)

df1
```
