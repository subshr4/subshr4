Here I imported pandas and then activity dataset by using read csv function. Then I created the first boxplot by importing seaborn as sns where x is equal to workout type and y is equal to calories. Its title is boxplot of calories burned. Also, 45 degree is used as rotation so all the workout type can be seen clearly. Then in the second boxplot I swaped the direction by changing x and y. In third box plot I changed the box color to green and also used sns.stripplot command to add the black jittered point in the box.

```python
import pandas as pa

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')

df 
```

```python
import seaborn as sns
ax = sns.boxplot(data = df, x = 'workout_type', y = 'calories')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 45)
ax.set_title('Box Plot of Calories Burned')






```

```python
sns.boxplot(data = df, y = 'workout_type', x = 'calories')
ax.set_title('Box Plot of Calories Burned')
```

```python
sns.boxplot(data = df, y = 'workout_type', x = 'calories', color = 'green')
sns.stripplot(data = df, y = 'workout_type', x = 'calories' , color = 'black')
ax.set_title('Box Plot of Calories Burned')
```
