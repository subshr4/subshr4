3. Here, I used Avg function to get the average time for a trip based on starting point (start_station_name), and (start_time) where the time is less the 2022-01-20.

```python
%%bigquery --project linear-skill-339004
SELECT 
    IFNULL(start_station_name, "") as start_station_name,
    TIMESTAMP_TRUNC(start_time, HOUR) as start_hour,
    AVG(duration_minutes) as average_time
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE DATE(start_time) < '2022-01-20'
GROUP BY start_station_name, start_time
ORDER BY start_station_name, start_time
```

4. Here, I used Count function to get the total number of trips based on each starting point (start_station_name), and (start_time) where the time is less the 2022-01-20.

```python
%%bigquery --project linear-skill-339004
SELECT 
    IFNULL(start_station_name, "") as start_station_name,
    TIMESTAMP_TRUNC(start_time, HOUR) as start_hour,
    COUNT(*) as num_trips
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE DATE(start_time) < '2022-01-20'
GROUP BY start_station_name, start_time
ORDER BY start_station_name, start_time
```
