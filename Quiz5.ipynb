 1. I created a new project (My First project) in the BigQuery. It's id is linear-skill-339004 which I have used to navigate to the dataset 'austin_bikeshare.bikeshare_trips'. Here I took this helpful starting code provided in the lecture 5. Here count is used to count total number of trips , and number of trips of each start_sation_name are ordered in a descending order.

```python
%%bigquery --project linear-skill-339004
SELECT 
  COUNT(*) as number_trips, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY number_trips DESC
```

1. Here I used not equal(!=) to the most popular trip to get the second most popular trip which is Riverside @ S. lamar.

```python

%%bigquery --project linear-skill-339004
SELECT start_station_name,
 COUNT(*) as most_popular_trip
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name

HAVING most_popular_trip!= (SELECT MAX(most_popular_trip)as max_trip
FROM (
SELECT COUNT(*) as most_popular_trip, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
))
ORDER BY most_popular_trip DESC
LIMIT 1
```

Here I took this helpful starting code provided in the lecture 5. Here, where symbol shows those station where start_ station is equal to end_station. Limit shows the number of data.

```python

%%bigquery --project linear-skill-339004
SELECT start_station_name, duration_minutes
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name = end_station_name
LIMIT 100
```

2. Here Countif function is used  to meet the requirement trips lasted over an hour and were a round trip.

```python
%%bigquery --project linear-skill-339004
SELECT *, COUNTIF(duration_minutes>60) as trip_over_an_hour
FROM (
  SELECT start_station_name, duration_minutes
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name = end_station_name
)
GROUP BY duration_minutes, start_station_name
Order BY trip_over_an_hour DESC
```
