```python


import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn')  # to get seaborn scatter plot

# read the csv file to extract data

data = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
distance = data['distance']
calories = data['calories']
workout_type = data['workout_type']


plt.scatter(distance, calories, s=30, alpha=0.6, linewidth=1)

plt.title('distance vs calories')
plt.xlabel('distance')
plt.ylabel('calories')

plt.show()
```

```python

plt.scatter(distance, calories, s=30, alpha=0.6, edgecolor='red',linewidth=1)

plt.title('distance vs calories')
plt.xlabel('distance')
plt.ylabel('calories')

plt.show()
```
