Here, I imported pandas, matplotlib.pyplot, and seaborn in order to create a graph. Then I imported activity dataset usind read_csv command. Here I have shown three different graphs. Firstly,I created histogram of time from activity dataset. Secondly, I graphed the average calories by workout_type, and plotted its bar graph. Lastly, I created a box plot of time from the same dataset.

```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df
```

```python
ax = df.time.hist(bins = 25)
ax.set(title = "Histogram of Time")
plt.show()
```

```python
dfgrouped = df.groupby('workout_type').calories.agg(['mean','std', 'count'])

dfgrouped.plot.bar(y = 'mean', color = ['green','red', 'blue'], title = "Error of Standard Deviation")
```

```python
df.time.plot.box()
```
