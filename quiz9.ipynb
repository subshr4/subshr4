Here, I imported pandas and beautifulsoup. Then I entered the URL of Marvel cinematic universe films to get its table.

```python
import requests
import pandas as pa
from bs4 import BeautifulSoup


r = requests.get('https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films')
html_contents = r.text
html_soup = BeautifulSoup(html_contents,"lxml")
```

Then, I entered the class name wikitable plainrowheader and find all function to list the table.

```python
tables = html_soup.find_all('table',class_="wikitable plainrowheaders")
tables[0].find_all('a')
```

 Here I passed it to pandas using the read_html command.

```python
df = pa.read_html(str(tables))[0]
df
```

There are two topics in each column so I used one droplevel(0) to remove the multi indexing that is going on table.

```python
df.columns = df.columns.droplevel(0)
df
```
