Firstly, I imported pandas dataframe and used read_csv to read the activitiy dataset.

(c) (10 points) Clean the column names to remove the unit declaration, (%), using regular expressions. The column name should not have any trailing spaces after cleaning. You will only receive partial credit for simply renaming columns without using regular expressions.

Here, I imported regular expression operation and  cleaned the unit declaration.

Reference: https://github.com/nurfnick/Data_Viz/blob/main/18_One_Hot_Encoding.ipynb

```python
import re
df = df.rename(columns=lambda x: re.sub(r'[(.%)]','',x))
```

```python
df.dtypes
```

(d) (10 points) Convert ‘activity day’ column into a datetime format.

Here, the datatype of activity_day is object, so in order to convert it into datetime, I used to_datetime function to convert it into datetime( %m/%d/%Y' ) format.




Reference: https://github.com/nurfnick/Data_Viz/blob/main/16_Dates.ipynb

```python
pa.to_datetime(df.activity_day, format='%m/%d/%Y')
```

```python
df.activity_day = pa.to_datetime(df.activity_day, format='%m/%d/%Y')
```

(e) (10 points) Impute ‘total steps’ by replacing the ‘NaN’s with an appropriate number of steps. Convert to appropriate datatype.


Here, I imported numpy and used fillna function to replace the null value with appropriate number of steps.

Reference: https://github.com/nurfnick/Data_Viz/blob/main/19_Missing_and_Incomplete.ipynb

```python
import numpy as na
df.fillna(value = 0)
```

(f) (10 points) Convert non-empty ‘avg pace’ into a float that is still representative of the information contained in the column. Recall that there are 60 seconds in one minute so 3:30 is equivalent to 3.5 minutes.


Here, I converted average pace into datetime and into a float.

```python
df['avg_pace'] = pa.DatetimeIndex(df['avg_pace'])
df['avg_pace'] = df['avg_pace'].apply(lambda avg_pace: avg_pace.hour + (avg_pace.minute / 60))
df['avg_pace'].head(6)
```

(g) (10 points) Group data by ‘workout type’ and find the mean, median, count and standard deviation of ‘calories’.

Here I grouped data by workout_type and found its mean, median, count, and std of calories by using agg function that passes the function.


Reference: https://github.com/nurfnick/Data_Viz/blob/main/20_Cleaning_Conclusion.ipynb

```python
df.groupby('workout_type').agg(['mean', 'median', 'count', 'std'])['calories']
```
